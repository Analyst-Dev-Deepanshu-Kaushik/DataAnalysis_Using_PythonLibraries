### **The analysis of the data will now be started.**

#### The total sale will first be determined.

```python
# The total sales are calculated and rounded to two decimal places
retail['TOT_SALES'].sum().round(2)    # Calculates and rounds the total sales to two decimal places
```

#### The total sale and average total sale by type of customer will now be determined.

```python
# The total and average sales are calculated for each customer type
retail.pivot_table(index=['PREMIUM_CUSTOMER'],values=['TOT_SALES'],aggfunc=['sum','mean'])    # Creates a pivot table to calculate sum and mean of total sales by customer type
```

#### It was found that Mainstream has the highest sale.

#### Next, the total sale and average total sale by lifestage or age group of customers will be determined.

```python
# The total and average sales are calculated for each lifestage group
retail.pivot_table(index=['LIFESTAGE'],values=['TOT_SALES'],aggfunc=['sum','mean'])    # Creates a pivot table to calculate sum and mean of total sales by lifestage
```

#### It was found that Older Singles/Couples have the highest sale, while New Families have the lowest sale.

#### Next, the total sale by age group and type of customer will be analyzed together.

```python
# The total sales are grouped by lifestage and customer type, then unstacked for better readability
retailgrp=retail.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].sum().unstack()    # Groups data by lifestage and customer type, then calculates total sales and reshapes the result
```

```python
# The grouped data showing total sales by lifestage and customer type is displayed
retailgrp    # Displays the grouped and reshaped total sales data by lifestage and customer type
```

#### A plot will now be created to better understand the result.

```python
# Set the size of the plot to make it more readable
plt.figure(figsize=(20,5))    # Sets the figure size to 20x5 inches for better visualization
# Create a stacked bar plot to display the total sales by customer lifestage and type
retailgrp.plot(kind='bar',stacked=True,color=['lightcoral','cyan','limegreen'])    # Plots a stacked bar chart with specified colors
# Set the title for the plot
plt.title('Total Sale by Customer Lifestage and Type of Customer')    # Adds a title to the plot
```

#### It was found from the plot that New Families have the lowest sale in each customer category.
#### Next, the total sale and average total sale by brand name will be analyzed.

```python
# Total and average sales are calculated for each brand
retail.pivot_table(index=['BRAND'],values=['TOT_SALES'],aggfunc=['sum','mean'])    # Creates a pivot table to calculate sum and mean of total sales by brand
```

#### It was found that Kettle has the highest sale, while Burger has the lowest sale.
#### A plot will be created to provide a better understanding of the sales by brand.

```python
# A bar plot is created to show the total sales by brand
sns.barplot(data=retail,x='BRAND',y='TOT_SALES',estimator=sum)    # Creates a bar plot to display the total sales for each brand
# The x-axis labels are rotated by 90 degrees for better readability
plt.xticks(rotation=90)    # Rotates the x-axis labels to improve readability
```

#### Create a bar plot using the Pandas plot function and GroupBy (with the same records and fields).

```python
# Group the data by brand and calculate the total sales for each brand
retailbrand=retail.groupby(['BRAND'])['TOT_SALES'].sum()    # Groups data by brand and calculates total sales
# Create a bar plot to display the total sales by brand
retailbrand.plot(kind="bar")    # Plots the total sales as a bar chart
# Set the title for the plot
plt.title("Total Sale by Brands")    # Adds a title to the bar plot
```

#### The difference between Kettle and Burger brands can be seen here.
#### Next, the total sale by brands and product names will be checked.

```python
# Group the data by product name and brand, and calculate the total sales for each combination
bpngrp=retail.groupby(['PROD_NAME','BRAND'])['TOT_SALES'].sum().unstack()    # Groups data by product name and brand, then calculates total sales
# Display the grouped total sales data by product name and brand
bpngrp    # Displays the total sales for each product and brand combination
```

#### The total sale by brands and product name will be checked using a pivot table (with the same fields).

```python
# Create a pivot table to calculate the total sales by brand and product name
bpnpvt=retail.pivot_table(index=['BRAND','PROD_NAME'],values=['TOT_SALES'],aggfunc=['sum'])    # Creates a pivot table showing the total sales for each brand and product name
# Display the pivot table showing the total sales by brand and product name
bpnpvt    # Displays the total sales data by brand and product name
```

#### It was found from the result that the product "Dorito Corn Chip Supreme 380g" has the highest sale, and "Woolworths Mild Salsa 300g" has the lowest sale.
### Next, the day, month, and year will be extracted from the DATE column for analysis.

```python
# Display the first few rows of the dataframe for inspection
retail.head()    # Displays the first 5 rows of the dataframe to review the data structure
```

#### The datatype of all the fields will be checked.

```python
# Display the data types of all the columns in the dataframe
retail.dtypes    # Shows the data types for each field in the dataframe
```

#### It can be observed that the DATE column has an Object datatype. Therefore, this column needs to be converted into Date format.

#### The DATE column will be converted into Datetime format.

```python
# Convert the 'DATE' column to Datetime format for proper date-related analysis
retail=retail.astype({'DATE':'datetime64[ns]'})    # Converts the 'DATE' column to Datetime format
# Display the data types to confirm the conversion
retail.dtypes    # Displays the data types of all columns to verify the 'DATE' column conversion
```

#### The month and year will now be extracted from the DATE column.

```python
# Extract the year and month from the 'DATE' column and store them in separate columns
retail['year']=retail['DATE'].dt.year    # Extracts the year from the 'DATE' column and stores it in the 'year' column
retail['month']=retail['DATE'].dt.month    # Extracts the month from the 'DATE' column and stores it in the 'month' column
```

```python
# Display the first few rows of the dataframe to verify the newly added 'year' and 'month' columns
retail.head()    # Displays the first 5 rows of the dataframe to confirm the addition of 'year' and 'month' columns
```

#### The total sale by year will be checked.

```python
# Create a pivot table to calculate the total sales for each year
retail.pivot_table(index=['year'],values=['TOT_SALES'],aggfunc=['sum'])    # Creates a pivot table to calculate the sum of total sales by year
```

#### It was found that 2018 (July to December) has higher sales than 2019 (January to June).
#### A deeper analysis of the year with months will be performed.

```python
# Create a pivot table to calculate the total sales for each year and month combination
retail.pivot_table(index=['year','month'],values=['TOT_SALES'],aggfunc=['sum'])    # Creates a pivot table to calculate the sum of total sales by year and month
```

#### It was found from the result that December has the highest sale, and February has the lowest sale.
#### A plot will be created to further understand the sales on a date-wise basis.

```python
# Set the figure size for better visualization of the plot
plt.figure(figsize=(25,10))    # Sets the figure size to 25x10 inches for better visibility
# Create a line plot to visualize the total sales over time (by date)
sns.lineplot(data=retail,x='DATE',y='TOT_SALES',estimator='sum')    # Plots a line chart showing the sum of total sales over time (date-wise)
```

#### The total sale by store numbers will be checked.

```python
# Create a pivot table to calculate the total sales for each store number
strsl=retail.pivot_table(index='STORE_NBR',values='TOT_SALES',aggfunc='sum')    # Creates a pivot table to calculate the total sales by store number
# Sort the pivot table by total sales in descending order to identify the top-performing stores
strsl.sort_values(by='TOT_SALES',ascending=False)    # Sorts the pivot table by total sales in descending order
```

#### It was found from this analysis that store number 226 has the highest sale, and store number 211 has the lowest sale.

#### The monthly sale for each store will now be checked.

```python
# Create a pivot table to calculate the total sales for each store number and month combination
retail.pivot_table(index=['STORE_NBR'],columns=['month'],values=['TOT_SALES'],aggfunc='sum')    # Creates a pivot table to calculate the sum of total sales by store number and month
```

#### It was found that store numbers 85, 92, and 76 have sales records for only one month.

#### The total sale by brands and pack size of products will now be checked.

```python
# Create a pivot table to calculate the total sales for each combination of brand and pack size
retail.pivot_table(index=['BRAND','PACK_SIZE'],values=['TOT_SALES'],aggfunc=['sum'])    # Calculates the sum of total sales for each brand and pack size combination
```
